### In this notebook i am creating a intermediate level of DA project using the S&P 500 Stocks (daily updated) dataset

### Link: [S&P 500 Stocks (daily updated)](https://www.kaggle.com/datasets/andrewmvd/sp-500-stocks)

```python
import pandas as pd
import os
print(os.listdir("../Data"))
```

```python
#Loading the Data
df = pd.read_csv('../Data/sp500_stocks.csv')

print(df.head())
print(df.info())
```

### Data cleaning and Wragling

```python
#Convert Date

df["Date"] = pd.to_datetime(df["Date"])
```

```python
#Check Missing Values
print(df.isnull().sum())
```

### Removing the rows with missing prices

```python
df= df.dropna(
    subset=["Open", "High", "Low", "Close"]
)

df
```

```python
#Fill the misssing volumes

df["Volume"] = df["Volume"].fillna(0)

df
```

### Ensuring the correct datatypes

```python
price_cols = ["Open", "High", "Low", "Close"]

df[price_cols] = df[price_cols].astype(float)

df["Volume"] = df["Volume"].astype(int)

df

```

### Filter Delisted Stocks

```python
latest_date = df["Date"].max()

active_stocks = (
    df.groupby("Symbol")["Date"].max()
)

active_stocks = active_stocks[active_stocks > latest_date - pd.Timedelta(days=30)].index

df = df[df["Symbol"].isin(active_stocks)]

df
```

### Feature Engineering

```python
df = df.sort_values(
    ["Symbol", "Date"]
)
```

### Daily Return Formula:

$$
\frac{P_t-P_{t-1}}{P_{t-1}}
$$

```python
df["Daily_Return"] = (
    df.groupby("Symbol")["Close"].pct_change()
)
```

### Percentage Change

```python
df["pct_change"] = (
    (df["Close"] - df["Open"])
    / df["Open"]
)* 100
```

### 7-Day Moving Average

```python
df["MA7"] = (
    df.groupby("Symbol")["Close"]
    .transform(
        lambda x: x.rolling(7).mean()
    )
)
```

### 30-Day Moving Average

```python
df["MA30"] = (
    df.groupby("Symbol")["Close"]
    .transform(
        lambda x: x.rolling(30).mean()
    )
)
```

### Volatality

```python
df["Volatality"] =(
    df.groupby("Symbol")["Daily_Return"]
    .transform(
        lambda x: x.rolling(30).std()
    )
)
```

### Volume Change

```python
df["Volume_Change"] = (
    df.groupby("Symbol")["Volume"]
    .pct_change()
)
```

### Exploratory Analysis

##### Which sectors Performed Best?

```python
stock_returns = (
    df.groupby("Symbol")["Daily_Return"]
    .mean()
    .sort_values(ascending= False)
)
print(stock_returns.head(10))
```

##### Which Stocks Were Most Volatile

```python
volatality = (
    df.groupby("Symbol")["Volatality"]
    .mean()
    .sort_values(ascending=False)
)
print(volatality.head(20))
```

##### What Did the 2020 COVID Crash Look Like?

```python
import matplotlib.pyplot as plt

crash_df = df[
    (df["Date"] >= "2020-02-01") &
    (df["Date"] <= "2020-02-29")
]

market = (
    crash_df.groupby("Date")
            ["Daily_Return"]
            .mean()
)

market.plot(figsize=(12,8))

plt.title("Average Market Return During COVID Crash")
plt.ylabel("Average Daily Return")

plt.show()
```

##### Which Stocks Recovered Fastest?

```python
recovery = []

for symbol in df["Symbol"].unique():
    stock = df[df["Symbol"] == symbol]
    low_price = stock["Close"].min()

    latest_price = stock["Close"].iloc[-1]

    recovery_pct = (
        (latest_price - low_price) / low_price
    )* 100

    recovery.append(
        [symbol, recovery_pct]
    )

recovery_df = pd.DataFrame(
    recovery,
    columns=["Symbol", "Recovery_%"]
)

print(
    recovery_df.sort_values(
        "Recovery_%", ascending=False
    ).head(10)
)
```

##### Do Higher Volumes Lead to Larger Price Changes?

```python
volume_price_corr = (
    df[["Volume_Change", "Daily_Return"]]
    .corr()
)

print(volume_price_corr)
```
