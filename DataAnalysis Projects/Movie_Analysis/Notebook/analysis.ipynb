### Now Checking for the missing values

```python
movies.isnull().sum().sort_values(ascending=False)
```

### Visualization

```python
plt.figure(figsize=(10,6))
movies.isnull().sum().sort_values(ascending=False).head(15).plot(kind='bar')
plt.title('Missing values')
plt.show()
```

### DataSet Columns

```python
movies.columns
```

```python
#Filtering the useful columns
[
    'budget',
    'generes',
    'popularity',
    'revenue',
    'runtime',
    'vote_average',
    'vote_count',
    'release_date'
]
```

### Now trying to analyze the revenue

```python
#List of top earning movies
top_revenue = movies[['title','revenue']].sort_values(
    by='revenue',
    ascending=False
).head(10)

top_revenue
```

```python
#Plotting the top earning movies
plt.figure(figsize=(10,8))

sns.barplot(
    x = 'revenue',
    y = 'title',
    data=top_revenue
)

plt.title('Top 10 Highest Grossing Movies')
plt.show()
```

### Now Analyzing the Budget

```python
plt.figure(figsize=(10,8))

sns.histplot(
    movies['budget'],
    bins=30,
    kde=True
)
plt.title("Budget Distribution")
plt.show()
```

### Movie Rating Analysis

```python
plt.figure(figsize=(10,8))

sns.histplot(
    movies['vote_average'],
    bins=20,
    kde=True
)

plt.title("Vote Average Distribution")
plt.show()
```

### Popularity Analysis

```python
#Top popular movies

movies[['title','popularity']]\
.sort_values(by='popularity',ascending=False,).head(10)
```

```python
#Visualization of top popular movies

top_pop = movies[['title','popularity']]\
.sort_values(by='popularity',ascending=False,).head(10)

plt.figure(figsize=(10,8))

sns.barplot(
    x ='popularity',
    y = 'title',
    data = top_pop
)
plt.title('Top 10 popular movies')
plt.show()
```

### Runtime Analysis

```python
#analysing the runtime of the movies

plt.figure(figsize=(10,8))

sns.histplot(
    movies['runtime'],
    bins=30,
    kde=True
)
plt.title('Movies Runtime Distribution')
plt.xlabel("Runtime(minutes)")
plt.show()
```
