### Basic Operations on numpy array


```python
import numpy as np

arr = np.arange(1,82)

print(arr)

matrix = arr.reshape(9,9)

print("\n Updated matrix \n",matrix)
```

```python
transposed = matrix.T

print("Transposed Matrix: \n", transposed)
```

```python
matrix1 = np.arange(1,82).reshape(9,9)
matrix2 = np.arange(82,163).reshape(9,9)
```

```python
print("Matrix1: \n", matrix1)
print("\nMatrix2: \n", matrix2)
```

### General Operations

```python
print("Sum")
print(matrix1 + matrix2)
print("Substract")
print(matrix1 - matrix2)
print("Divide")
print(matrix1 / matrix2)
print("Multiply")
print(matrix1 * matrix2)
```

```python
#printing specific rows 
row_1 = matrix[0]
row_2 = matrix[1]
print(row_1,"\n", row_2)
```

```python
#printing specific columns 
col_1 = matrix[:,0]
col_2 = matrix[:,1]
print(col_1,"\n", col_2)
```

```python
#slicing 
# First two rows
print(matrix[:2, :])
print('\n')
# Last two columns
print(matrix[:, 1:])
```

```python
#accessing a single element 

print(matrix[0,8])
```
