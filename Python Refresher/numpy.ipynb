### Doing some experiment with np array

```python
import numpy as np
```

```python
array = np.array([20,10,40,90,80,50,70,30]) 
```

```python
#sum of the list
sum = sum(array)
print(sum)

#max of the list
print(max(array))

#min of the list
print(min(array))

#sort in ascending order
sort_nums = sorted(array, reverse=False)
print(sort_nums)

#sort in descending order
sort_nums = sorted(array, reverse=True)
print(sort_nums)

#printing first 3 values
print(array[:3])

#printing last 3 values
print(array[-3:])

#printing every second element
print(array[::2])
```
