```python
import pandas as pd
df = pd.read_csv("./Dataset/iris.csv")
df.head()
```

```python
inputs = df.drop("species", axis="columns")
target = df["species"]

inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LabelEncoder()

le_sepal_length = LabelEncoder()
le_sepal_width = LabelEncoder()
le_petal_length = LabelEncoder()
le_petal_width = LabelEncoder()
```

```python
inputs['sepal_length_'] = pd.Series(
	le_sepal_length.fit_transform(inputs['sepal_length']),
	index=inputs.index,
	dtype='int64'
)
inputs['sepal_width_'] = pd.Series(
	le_sepal_width.fit_transform(inputs['sepal_width']),
	index=inputs.index,
	dtype='int64'
)
inputs['petal_length_'] = pd.Series(
	le_petal_length.fit_transform(inputs['petal_length']),
	index=inputs.index,
	dtype='int64'
)
inputs['petal_width_'] = pd.Series(
	le_petal_width.fit_transform(inputs['petal_width']),
	index=inputs.index,
	dtype='int64'
)


inputs.head()
```

```python
input_n = inputs.drop(['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], axis='columns')

input_n
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split


model = tree.DecisionTreeClassifier()
```
