### This is a notebook where i am trying some experiments on data preporcessing to enhance my understandability of data pre-processing.

##### 3 main strategies
1 Drop — remove rows or columns with too many nulls. Simple but loses data.

2 Fill with a statistic — use mean/median for numbers, mode for categories.

3 Predict (imputation) — use another model to estimate the missing value.

```python
import pandas as pd
from sklearn.impute import SimpleImputer
```

## Performing operations on missing values

```python
df = pd.read_csv('./Datasets/Advertising Budget and Sales.csv')

df.head()
```

```python
# Detect missing values
print(df.isnull().sum())
```

```python
# Strategty 1: drop rows with any null
df_dropped = df.dropna()
df_dropped
```

```python
# Strategy 2: fill with median
imputer = SimpleImputer(strategy='median')
df['Sales ($)'] = imputer.fit_transform(df[['Sales ($)']])
df
```

## Working on duplicates

```python
#How Many duplicates
print(df.duplicated().sum())
df
```

```python
# See the duplicate rows
print(df[df.duplicated(keep=False)])
```

```python
# Remove exact duplicates (keep first)
df = df.drop_duplicates()
df
```

```python
# Remove duplicates by specific columns
df = df.drop_duplicates(subset=['Radio Ad Budget ($)'])
```
