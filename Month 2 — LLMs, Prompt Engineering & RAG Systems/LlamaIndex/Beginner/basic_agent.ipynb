# Hi I am trying to implement a basic agent using LLamaIndex

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from llama_index.core.agent.workflow import FunctionAgent
from llama_index.llms.ollama import Ollama

def get_information() -> str:
    """ Retrieves the detailed specifications, engine options, interior features, 
    and safety configurations of the Maruti Suzuki Fronx crossover SUV.
    
    Returns:
        str: A comprehensive text breakdown of the Fronx specifications.
    """
    return (
        "The Maruti Suzuki Fronx is a compact crossover SUV built on the Suzuki HEARTECT platform, "
        "balancing modern performance with structural rigidity. It offers two main petrol engine configurations: "
        "a refined 1.2L K-Series Dual Jet engine generating 88.7 bhp and 113 Nm of torque, and a punchier 1.0L Turbo "
        "Boosterjet engine equipped with progressive smart hybrid technology that produces 99 bhp and 147.6 Nm of torque. "
        "A highly efficient factory-fitted CNG option is also available on the 1.2L engine, dropping performance slightly to "
        "76 bhp but maximizing fuel economy. Depending on the engine selection, transmission options include a 5-speed manual, "
        "a 5-speed Automated Manual Transmission (AMT), or a smooth 6-speed torque-converter automatic complete with paddle shifters. "
        "Designed to handle diverse road conditions, the vehicle features a commanding 190 mm of ground clearance, riding on 16-inch "
        "dual-tone alloy wheels with a footprint measuring 3,995 mm in length, 1,765 mm in width, and a 2,520 mm wheelbase. Inside its "
        "dual-tone cabin, the premium trims come heavily equipped with a 9-inch SmartPlay Pro+ touchscreen infotainment system, wireless "
        "smartphone connectivity, a 360-degree surround-view camera, wireless charging, and a pop-up Heads-Up Display (HUD). Safety is also "
        "a major focus, as the crossover offers six airbags, Electronic Stability Program (ESP), and hill-hold assist across its model range."
    )
```

```python
llm = Ollama(model="qwen2.5:3b", request_timeout=120.0)

agent = FunctionAgent(
    llm= llm,
    tools=[get_information],
    system_prompt="You are an helpful assistant"

)
```

```python
response = await agent.run("What is the tyre size of fronx")
```
