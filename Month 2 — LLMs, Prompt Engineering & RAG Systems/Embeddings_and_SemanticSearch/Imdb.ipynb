```python
import pandas as pd
from sentence_transformers import SentenceTransformer

```

```python
df = pd.read_csv("./Data/imdb_top_1000.csv").head(100)  # start small
model = SentenceTransformer("all-MiniLM-L6-v2")

```

```python
embeddings = model.encode(df["Overview"].tolist(), show_progress_bar=True)
print(f"Embedded {len(embeddings)} reviews, shape: {embeddings.shape}")
for idx, embedding in enumerate(embeddings, start=1):
    print(f"Embedding {idx}: {embedding}")
```
