### The goal is to create a searchable index of Amazon reviews using embeddings.

```python
import pandas as pd

df = pd.read_csv(".//Dataset/Reviews.csv")
```

```python
df.head(10)
```

### Keep only the review text

```python
reviews = df["Text"].head(1000).tolist()
```

### Choose an embedding model

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
```

### Generate embeddings

```python
embeddings = model.encode(
    reviews,
    show_progress_bar=True
)
```

### Save Embeddings

```python
import numpy as np

np.save("review_embeddings.npy", embeddings)
```

```python
embeddings = np.load("review_embeddings.npy")
```

### Embed the query

```python
query = "Best Energy Drink"

query_embedding = model.encode([query])
```

### Compute cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

scores = cosine_similarity(query_embedding, embeddings)[0]
```

### Rank Results

```python
indices = np.argsort(-scores)
```

```python
for i in indices[:5]:
    print(reviews[i], "\n")
```
