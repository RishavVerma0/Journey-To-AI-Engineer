## Initializing and invoking a model

```python
from langchain.chat_models import init_chat_model
model = init_chat_model(model="ollama:qwen2.5:3b")
```

```python
response = model.invoke("What's the capital of the Saturn?")
response
```

```python
print(response.content)
```

```python
from pprint import pprint

pprint(response.response_metadata)
```

## Customising your Model

```python
model = init_chat_model(
    model="ollama:qwen2.5:3b",
    temperature=1.0
)

response = model.invoke("What's the capital of the Saturn?")
response
```

## Initialising and invoking an agent

```python
from langchain.agents import create_agent

agent = create_agent(model=model)
```

```python
from langchain.messages import HumanMessage

response = agent.invoke(
    {"messages": [HumanMessage(content="What's the capital of the Saturn?")]}
)
```

```python
pprint(response)
```

```python
print(response['messages'][-1].content)
```

```python
from langchain.messages import AIMessage

response = agent.invoke(
    {"messages": [HumanMessage(content="What's the capital of the Saturn?"),
    AIMessage(content="The capital of the Moon is Sola City."),
    HumanMessage(content="Interesting, tell me more about Sola City")]}
)

pprint(response)
```

## Streaming Output

```python
for token, metadata in agent.stream(
    {"messages": [HumanMessage(content="Tell me all about Sola City, the capital of Saturn")]},
    stream_mode="messages"
):
    # token is a message chunk with token content
    # metadata contains which node produced the token
    
    if token.content:  # Check if there's actual content
        print(token.content, end="", flush=True)  # Print token
```
