# No memory

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain.agents import create_agent

agent = create_agent(
    model="ollama:qwen2.5:3b"
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="Hello! my name is Rishav and my favorite car is Suzuki Fronx")

response = agent.invoke(
    {"messages":[question]}
)

print(response['messages'][-1].content)
```

```python
from pprint import pprint
pprint(response)
```

```python
question = HumanMessage(content="What is the name of my favorite car?")

response = agent.invoke(
    {"messages": [question]}
)
pprint(response['messages'][-1].content)
```

# Memory

```python
from langgraph.checkpoint.memory import InMemorySaver

agent = create_agent(
    model="ollama:qwen2.5:3b",
    checkpointer=InMemorySaver(),
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="Hello! my name is Rishav and my favorite car is Suzuki Fronx")
config = {"configurable": {"thread_id": "1"}}

response = agent.invoke(
    {"messages": [question]},
    config,  
)
pprint(response)
```

```python
question = HumanMessage(content="What is the name of my favorite car?")

response = agent.invoke(
    {"messages": [question]},
    config,
)

pprint(response)
```
