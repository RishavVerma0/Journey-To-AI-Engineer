### Daily dataset index — load Wikipedia intro paragraphs

```python
import wikipedia
import chromadb
def fetch_intro(title):
    try:
        return wikipedia.summary(title, sentences=3)
    except Exception as e:
        print(f"Skipped {title}: {e}")
        return None

topics = [
    "Python (programming language)", "Machine learning", "Artificial intelligence",
    "ChromaDB", "Natural language processing", "Data science",
    # ... add as many as you want, or pull a random/trending list daily
]

wiki_docs, wiki_ids, wiki_metas = [], [], []
for i, topic in enumerate(topics):
    intro = fetch_intro(topic)
    if intro:
        wiki_docs.append(intro)
        wiki_ids.append(f"wiki_{i}")
        wiki_metas.append({"project": "daily_dataset_index", "title": topic, "source": "wikipedia"})

# Create a persistent client
client = chromadb.PersistentClient(path="./chroma_db")

collection = client.get_or_create_collection("wikipedia_docs")

collection.add(
    documents=wiki_docs,
    metadatas=wiki_metas,
    ids=wiki_ids
)

print(f"Added {len(wiki_docs)} Wikipedia intros")
```
